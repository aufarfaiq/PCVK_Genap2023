# **MEMBUAT GARIS**

### **HORIZONTAL**

```python
image_1 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#Membuat garis horizontal di tengah image
for y in range (image.shape[1]):
  image_1[int(image.shape[0]/2),y] = [255, 255, 0]
cv2_imshow(image_1)
```

### **VERTIKAL**

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#Membuat garis horizontal di tengah image
for x in range (image.shape[0]):
  image_2[x,int(image.shape[1]/2)] = [0, 255, 255]
cv2_imshow(image_2)
```

#**FLIP**

### **VERTIKAL**

```python
image_flipV = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)
plt.imshow(image_flipV)
```

### **HORIZONTAL**

```python
image_flipH = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),1)
plt.imshow(image_flipH)
```

# **MENAMBAH SHAPE**

```python
img = cv.imread('/content/drive/MyDrive/soal3.jpg')

cv.rectangle(img,pt1=(60,60),pt2=(192,150),color=(0,255,255),thickness=5)
cv.rectangle(img,pt1=(60,100),pt2=(11,120),color=(0,255,255),thickness=5)
cv.rectangle(img,pt1=(240,100),pt2=(192,120),color=(0,255,255),thickness=5)
cv.rectangle(img,pt1=(310,80),pt2=(370,150),color=(255,0,0),thickness=5)
cv.rectangle(img,pt1=(310,100),pt2=(255,120),color=(255,0,0),thickness=5)
cv.circle(img,center=(608,150),radius=10,color=(255,0,255),thickness=20)
plt.imshow(img)
```
